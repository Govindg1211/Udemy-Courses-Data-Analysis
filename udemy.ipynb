## Data Preprocessing 

### Data Cleaning

#### Handling Null Values

```python
# Dropping these ['price_detail__price_string','price_detail__currency'],['discount_price__price_string','discount_price__currency'] columns
# ['price_detail__amount'],['discount_price__amount'] have same data as of these column ['price_detail__price_string'],['discount_price__price_string'] respectively. 
# In Integer Datatype. 
# ['avg_rating_recent','avg_rating,rating'] these three column cotaining similar data. Therefore removing two column. ( Float Data Type)

df = df.drop(columns=['discount_price__currency' , 'discount_price__price_string' , 'price_detail__price_string' , 'price_detail__currency'])
df = df.drop(columns=['avg_rating','rating'])
```

```python
# Filling Null value as 0 in this column ['discount_price__amount']
df['discount_price__amount'].fillna(0,inplace=True)
df['price_detail__amount'].fillna(0,inplace=True)
```

### Feature  Engineering

#### Feature Transformation

```python
# Converting these columns ['created','published_time'] into Data Time Format
df["created"] = pd.to_datetime(df["created"], errors='coerce')
df["published_time"] = pd.to_datetime(df["published_time"], errors='coerce')
```

```python
# Extract the year from the ['published_time'] column
df["published_year"] = df["published_time"].dt.year
# Display the first few rows to verify
df[["created", "published_time", "published_year"]].head()
```

```python
# Converting Data type of ['is_paid'] from Boolean to integer
df[['is_paid']] = df[['is_paid']].astype(int)
```

#### Dropping Unwanted Columns 

```python
# Dropping ['created','published_time'] these columns
df = df.drop(columns=['created','published_time'])

# Dropping Irrelevant Columns
df = df.drop(columns=['id','title','url','is_wishlisted'])
```

## Model Building

```python
import joblib
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score, KFold
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor, StackingRegressor, ExtraTreesRegressor
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.svm import SVR
from xgboost import XGBRegressor
from catboost import CatBoostRegressor
from lightgbm import LGBMRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.tree import DecisionTreeRegressor
import numpy as np

# Feature Engineering
df["subscribers_per_review"] = df["num_subscribers"] / (df["num_reviews"] + 1)
df["lecture_density"] = df["num_published_lectures"] / (df["num_subscribers"] + 1)
df["discount_percentage"] = (df["price_detail__amount"] - df["discount_price__amount"]) / df["price_detail__amount"]
df.drop(columns=["price_detail__amount", "num_published_practice_tests"], inplace=True)
df.fillna(0, inplace=True)

# Define Features & Targets
X = df.drop(columns=["num_subscribers", "avg_rating_recent"])  # Features
y_subscribers = df["num_subscribers"]
y_rating = df["avg_rating_recent"]

# Standardization
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA (retain 95% variance)
pca = PCA(n_components=0.99)
X_pca = pca.fit_transform(X)
joblib.dump(scaler, "scaler.pkl")
joblib.dump(pca, "pca.pkl")

# Train-Test Split
X_train, X_test, y_train_sub, y_test_sub = train_test_split(X_pca, y_subscribers, test_size=0.2, random_state=42)
X_train, X_test, y_train_rate, y_test_rate = train_test_split(X_pca, y_rating, test_size=0.2, random_state=42)

# Define Base Models for Stacking
base_models = [
    ("rf", RandomForestRegressor(n_estimators=200, max_depth=12)),
    ("dt", DecisionTreeRegressor(max_depth=12)),
    ("xgb", XGBRegressor(n_estimators=200, learning_rate=0.1, eval_metric="rmse")),
    ("cat", CatBoostRegressor(verbose=0)),
    ("extratrees", ExtraTreesRegressor(n_estimators=150))
]

# Meta-Model (Final Model)
meta_model = MLPRegressor(hidden_layer_sizes=(50, 25), max_iter=500, random_state=42)

# Stacking Regressor
stacking_regressor_subscribers = StackingRegressor(estimators=base_models, final_estimator=meta_model)
stacking_regressor_rating = StackingRegressor(estimators=base_models, final_estimator=meta_model)

# Train Models
stacking_regressor_subscribers.fit(X_train, y_train_sub)
stacking_regressor_rating.fit(X_train, y_train_rate)

# Save models
joblib.dump(stacking_regressor_subscribers, "model_subscribers.pkl")
joblib.dump(stacking_regressor_rating, "model_rating.pkl")

# Predictions
y_train_pred_sub = stacking_regressor_subscribers.predict(X_train)
y_test_pred_sub = stacking_regressor_subscribers.predict(X_test)

y_train_pred_rate = stacking_regressor_rating.predict(X_train)
y_test_pred_rate = stacking_regressor_rating.predict(X_test)

# Cross-validation settings
cv = KFold(n_splits=5, shuffle=True, random_state=42)

# Evaluate Models
def evaluate_model(y_train, y_test, y_train_pred, y_test_pred, model, model_name):
    r2_train = r2_score(y_train, y_train_pred)
    r2_test = r2_score(y_test, y_test_pred)
    
    # Cross-validation R² score
    cv_scores = cross_val_score(model, X_train, y_train, cv=cv, scoring='r2')
    r2_cv_mean = np.mean(cv_scores)

    print(f"\n📌 Performance for {model_name}:")
    print(f"Train R²: {r2_train:.4f}, Test R²: {r2_test:.4f}, Cross-Validation R²: {r2_cv_mean:.4f}")
    print(f"MAE: {mean_absolute_error(y_test, y_test_pred):.4f}, MSE: {mean_squared_error(y_test, y_test_pred):.4f}")
    
# Evaluate for `num_subscribers`
evaluate_model(y_train_sub, y_test_sub, y_train_pred_sub, y_test_pred_sub, stacking_regressor_subscribers, "num_subscribers")

# Evaluate for `avg_rating_recent`
evaluate_model(y_train_rate, y_test_rate, y_train_pred_rate, y_test_pred_rate, stacking_regressor_rating, "avg_rating_recent")
```

```python
def predict_sample(data):
    # Load saved models
    scaler = joblib.load("scaler.pkl")
    pca = joblib.load("pca.pkl")
    model_sub = joblib.load("model_subscribers.pkl")
    model_rate = joblib.load("model_rating.pkl")

    # Preprocess data
    data_scaled = scaler.transform(data)
    data_pca = pca.transform(data_scaled)

    # Make predictions
    pred_subscribers = model_sub.predict(data_pca)
    pred_rating = model_rate.predict(data_pca)
    return pred_subscribers, pred_rating

# Example sample data
sample_data = np.array([[5, 10, 3.5, 20, 2, 15, 7, 0.5]]) 
pred_sub, pred_rate = predict_sample(sample_data)
print(f"📌 Predicted Subscribers: {pred_sub[0]:.2f}")
print(f"📌 Predicted Avg Rating: {pred_rate[0]:.2f}")
```
